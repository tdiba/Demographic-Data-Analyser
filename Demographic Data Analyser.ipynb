## **Question 3: Percentage with Bachelor’s degree**

```python
percentage_bachelors = round(
    (df[df["education"] == "Bachelors"].shape[0] / df.shape[0]) * 100, 1
)
percentage_bachelors
```

## **Questions 4 and 5: Advanced education vs not advanced education**

#### **Defining Advanced Education**

```python
higher_education = df["education"].isin(["Bachelors", "Masters", "Doctorate"])
lower_education = ~higher_education
```

#### **Finding rich people in both groups**

```python
higher_education_rich = df[higher_education & (df["salary"] == ">50K")]

lower_education_rich = df[lower_education & (df["salary"] == ">50K")]
```

#### **Calculating percentages**

```python
higher_education_rich = round(
    (higher_education_rich.shape[0] / df[higher_education].shape[0]) * 100, 1
)

lower_education_rich = round(
    (lower_education_rich.shape[0] / df[lower_education].shape[0]) * 100, 1
)
```

## **Question 6: Minimum hours worked per week**

```python
min_work_hours = df["hours-per-week"].min()
min_work_hours
```

## **Question 7: Percentage of rich people among minimum-hour workers**

```python
num_min_workers = df[df["hours-per-week"] == min_work_hours] #filtering people who work the minimum hours
rich_min_workers = num_min_workers[num_min_workers["salary"] == ">50K"] #from that group, filter those who earn >50K
rich_min_workers
```

```python
rich_percentage = round(
    (rich_min_workers.shape[0] / num_min_workers.shape[0]) * 100, 1
)
rich_percentage
```

## **Question 8: Country with highest percentage of people earning >50K**

```python
country_count = df["native-country"].value_counts() #count total people in each country
```

```python
country_rich_count = df[df["salary"] == ">50K"]["native-country"].value_counts() #filter people earning more than 50k adn filter by country
```

```python
country_rich_percentage = (country_rich_count / country_count) * 100 #Calculate rich percentage per country
```

```python
#finding the highest earning country
highest_earning_country = country_rich_percentage.idxmax() 
highest_earning_country_percentage = round(country_rich_percentage.max(), 1) 
highest_earning_country
```
